Для минимизации функций я использую пакет `cvxpy`.

```python
import numpy as np
import sys
sys.path.append('/usr/local/lib/python2.7/dist-packages/')
import cvxpy
import matplotlib.pyplot as plt
```

Функция чтения из файла. Из-за неоднородности данных в файле (вектор $b$ записан отдельно) приходится писать отдельную функцию.

```python
def read_file(file_name, num_measuring):
    A, b = [], []
    line_index = 0
    with open(file_name) as file:
        while line_index < num_measuring:
            A += [np.array(file.readline().split('  ')[1:], dtype=float)]
            line_index += 1
        while line_index < 2 * num_measuring:
            b += [float(file.readline()[:-2])]
            line_index += 1
    return np.array(A), np.array(b)
```

Я пробовал три разных подхода:
$$||Ax - b||_2^2 \to \min_x$$
$$||Ax - b||_2^2 + ||x||_2^2 \to \min_x$$
$$||Ax - b||_2^2 + ||x||_1 \to \min_x$$

Второй способ оказался лучшим судя по результатам.
Третий способ, кажется, хуже всего.
Введение констант при коэффициентах в регуляризации не особо повлияло на результат, поэтому привожу результат без констант.
Можно так же заметить, что во всех случаях в матрице $A^T A$ некоторые собственные значения равны нулю.

```python
def find_area(file_path, m):
    print 'Working with file', file_path
    A, b = read_file(file_path, m)
    A = np.matrix(A)
    dim = sqrt(A.shape[1])
    
    print 'Number of lines:', m
    print 'Dimention:', dim
    
    eigenvalues = np.linalg.eig(A.T * A)[0]
    print 'Eigen values:', eigenvalues
    
    x = cvxpy.Variable(rows=A.shape[1])
    exprs = [cvxpy.norm(A * x - b) ** 2, 
            cvxpy.norm(A * x - b) ** 2 + cvxpy.norm(x, 2) ** 2,
            cvxpy.norm(A * x - b) ** 2 + cvxpy.norm(x, 1)]

    labels = ['without reg', '2 reg', '1 reg']
    plt.figure(figsize=(15, 5))

    print 'Minimums of expressions:'
    for i, expr in enumerate(exprs):
        problem = cvxpy.Problem(cvxpy.Minimize(expr))
        print problem.solve(), '-', labels[i]
        picture = x.value.reshape((dim, dim))

        plt.subplot(1, 3, i + 1)
        plt.imshow(1 - picture, cmap='gray')
        plt.axis('off')
        plt.title(labels[i])

    plt.show()
```

Лучший результат дает 2-регуляризация.

```python
find_area('set1.txt', 300)
find_area('set2.txt', 400)
find_area('set3.txt', 400)
```
