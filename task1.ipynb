## Применяем PCA для выборки профиль и давление

Найдем средние значения и ковариационную матрицу

```python
yp_mean = [mean(v) for v in transpose(sample_train)]
y_mean = yp_mean[:57]
p_mean = yp_mean[57:]
cov_matr = cov(transpose(sample_train))
cov_matr.shape
```

Построим PCA и посчитаем на тестовой выборке для каждого $d$ ошибку по формуле
$$err = \sqrt{ \frac {1} {M} \sum_{i=1}^M \left|\left| P_i - P_i^* \right|\right| ^ 2 }$$

```python
error = zeros(50)
for d in range(1, 51):
    pca = PCA(n_components=d)
    pca.fit(cov_matr)
    eigenvector = pca.components_
    y_eigenvector = transpose(matrix(eigenvector[:, :57]))
    p_eigenvector = matrix(eigenvector[:, 57:])

    U, s, V = np.linalg.svd(y_eigenvector, full_matrices=True)

    S_obr = np.zeros(y_eigenvector.shape, dtype=double)
    S_obr[:d, :d] = np.diag(s ** (-1))
    S_obr = transpose(S_obr)

    pseudoinverse = matrix(transpose(V)) * matrix(S_obr) * matrix(transpose(U))

    lambd = pseudoinverse * transpose(matrix(sample_test[:, :57] - y_mean))
    predict_p = transpose(lambd) * p_eigenvector + p_mean
    error_vector = [sum((array(predict_p[i]) - sample_test[i, 57:]) ** 2) for i in range(len(sample_test))]
    error[d-1] = sqrt(sum(error_vector) / len(sample_test))
```

Построим график ошибки в зависимости от $d$

```python
plt.figure(figsize=(18,6))
plt.plot(range(1, 51), error, 'ro-')
plt.show()
```

Видим, что наименьшая ошибка получается при $d=3$. То есть для того, чтобы наилучшим образом восстановить давление, достаточно только три главные компоненты. Нарисуем несколько графиков для $d = 3$. Синим показано истинное давление, красным --- восстановленное по профилю.

```python
d = 3

pca = PCA(n_components=d)
pca.fit(cov_matr)
eigenvector = pca.components_
y_eigenvector = transpose(matrix(eigenvector[:, :57]))
p_eigenvector = matrix(eigenvector[:, 57:])

U, s, V = np.linalg.svd(y_eigenvector, full_matrices=True)

S_obr = np.zeros(y_eigenvector.shape, dtype=double)
S_obr[:d, :d] = np.diag(s ** (-1))
S_obr = transpose(S_obr)

pseudoinverse = matrix(transpose(V)) * matrix(S_obr) * matrix(transpose(U))

lambd = pseudoinverse * transpose(matrix(sample_test[:, :57] - y_mean))
predict_p = transpose(lambd) * p_eigenvector + p_mean

for i in range(10):
    plt.figure(figsize=(20,3))
    plt.plot(range(57), sample_test[i, 57:], 'bo-')
    plt.plot(range(57), array(predict_p)[i], 'ro-')
    plt.show()
```
