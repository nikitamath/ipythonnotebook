```python
import numpy as np
import scipy.stats as sps
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

$$\xi \sim N \left( 0,  
\begin{pmatrix} 
        10 & 8\\
 	 8 & 10 \\
\end{pmatrix} \right)$$

```python
mean = [0, 0]
cov = [[10, 8], [8, 10]]
x, y = np.mgrid[-10:10:0.05, -10:10:0.05]
pos = np.empty(x.shape + (2,))
pos[:, :, 0] = x; pos[:, :, 1] = y
z = sps.multivariate_normal.pdf(pos, mean=mean, cov=cov)
```

```python
plt.figure(figsize=(6, 6))
CS = plt.contour(x, y, z, [0.0264, 0.025, 0.018, 0.01, 0.005, 0.001])
plt.clabel(CS, fontsize=8, inline=1)
plt.xlabel('x1'), plt.ylabel('x2')
plt.show()
```

```python
plt.figure(figsize=(6, 6))
plt.imshow(z[::-1])
plt.axis('off')
plt.show()
```

```python
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, color='green', alpha=0.1)
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
plt.show()
```

Графики условной плотности $f_{\xi_2 | \xi_1}$

```python
x = np.linspace(-10, 10, 1000)
plt.figure(figsize=(15, 6))
for xi_1 in [-10, -8, -3, 0, 2, 5, 9]:
    y = sps.norm.pdf(x, loc=0.8 * xi_1, scale=0.6)
    plt.plot(x, y, label='$\\xi_1$ = {}'.format(xi_1))
plt.legend()
plt.show()
```
