# Задача 2

Исходные данные

```python
x = np.linspace(0.0, 2.0, 21)
print x
```

```python
X = transpose(matrix([1 + zeros(21), x, x**2, x**3, x**4, x**5]))
```

```python
y = matrix([-0.3362,1.806,2.8101,0.6112,2.2959,2.8481,0.1808,0.5416,2.793,2.0963,3.5,
           3.9834,4.2866,5.3261,5.2257,6.3355,4.6026,6.5253,7.2156,6.6904,9.5773])
print y
```

Подсчитаем для разных $\lambda$ ошибку по методу скользящего контроля и по формуле из задачи 1. Сравним эти значения между собой. Как видим, они всегда совпадают.

```python
S_for_lambda = []
Lambdas = np.linspace(0.1, 15.0, 1500)
for lambd in Lambdas:
    H = X * (transpose(X) * X + lambd * eye(6)) ** (-1) * transpose(X)
    t = H *  transpose(y)
    h = diag(H)
    S_formula = sum([((y[0, i] - t[i, 0]) / (1 - h[i])) ** 2 for i in range(21)]) / 21
    S_for_lambda += [S_formula]
    S = 0
    for i in range(21):
        indexs = range(i) + range(i + 1, 21)
        y_i = y[0, indexs]
        X_i = X[indexs, :]
        S += (y[0, i] - X[i, :] * (transpose(X_i) * X_i + lambd * eye(6)) ** (-1) 
              * transpose(X_i) * transpose(y_i)) ** 2
    S = S / 21
    if abs(S - S_formula) > 10 ** (-8):
        print 'error: lambda =', lambd, ' S =', S, 'S_formula =',S_formula
```

Оптимальное значение $\lambda$

```python
lambda_optim = min(zip(S_for_lambda, Lambdas))[1]
lambda_optim
```

График величины $S^*$ в зависимости от $\lambda$

```python
plt.figure(figsize=(17,6))
plt.plot(Lambdas, S_for_lambda, 'r.')
plt.show()
```

Построим регрессию для полученного $\lambda_{optim}$

```python
x_new = np.linspace(-0.1, 2.1, 300)
X_new = transpose(matrix([1 + zeros(300), x_new, x_new**2, x_new**3, x_new**4, x_new**5]))
Y_new = X_new * (transpose(X) * X + lambda_optim * eye(6)) ** (-1) * transpose(X) * transpose(y)
```

```python
plt.figure(figsize=(17,6))
plt.plot(x_new, Y_new, 'g-')
plt.plot(x, transpose(y), 'bo')
plt.xlim((-0.1, 2.1))
plt.show()
```

Также построим регрессию для двух других значений $\lambda$

```python
X = transpose(matrix([1 + zeros(7), array(x), array(x)**2]))
Y = matrix(y)
```

Посчитаем для равномерной сетки на отрезке [-4, 4] разбиением на 20 000 точек

```python
p = 0.95
values = []
x_new = x + [0]
X_new = transpose(matrix([1 + zeros(8), array(x_new), array(x_new)**2]))
for y_0 in np.linspace(-4, 4, 20000):
    y_new = matrix(y + [y_0])
    Y_new = X_new * (transpose(X_new) * X_new) ** (-1) * transpose(X_new) * transpose(y_new)
    m = sorted(zip(sorted(zip(array(abs(transpose(Y_new) - y_new))[0], range(8))), range(8)),
           key=lambda x: x[0][1])[7][1] + 1
    if m / 8. < 0.95:
        values += [y_0]
```

Точки выборки отмечены синим. Доверительное множество отмечено зеленым.

```python
figure(figsize=(17,6))
plt.plot(x, transpose(y), 'bo')
plt.plot(zeros(len(values)), values, 'g-', linewidth=3)
plt.xlim((-1.1, 1.1))
plt.ylim((-0.3, 1.3))
plt.show()
```
