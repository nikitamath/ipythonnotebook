Вспомогательная функция

```python
from scipy.stats import gaussian_kde
def den_plot(data, c = ['blue'], l = 18, w = 2.5):
    fig = figure(figsize=(l, w)), grid(True)
    for i in range(len(data)):
        density = gaussian_kde(data[i])
        xs = linspace(min(data[i]), max(data[i]), 200)
        plot(xs, density(xs), linewidth=2.5, color = c[i])
        scatter(data[i], zeros(len(data[i])), s=40, alpha=0.2, color = c[i])
    return fig
```

### Итоги

Получились такие правила:

жен, 1 или 2 класс => 1

возр < 18, 1 или 2 класс => 1

возр > 15, 3 класс => 0

муж, 2 или 3 класс => 0

муж, 1 класс, возр > 42 => 0

муж, 17 < возр < 43, 1 класс => rand

возр < 16, 3 класс => rand

Метка 1 означает, что пассажир выживает, а метка 0 --- нет. Метка rand означает, что по таким критериям ничего конкретного сказать нельзя, то есть можно выдавать случайно 0 или 1. На самом деле, во-первых, такой путь не имеет преимуществ, чем если в таких случаях выдавать одну из констант. Дествительно, вероятность совпадения двух независимых бернуллиевских случайных величин с параметром 0.5 равна 0.5. То же значение вероятности можно получить, рассматривая события равенства одной из них константе. Поэтому вероятность ошибки в обоих случаях будет 0.5. Во-вторых, выдавать случайные ответы может быть опасно, поскольку может случится, что ошибки будут случаться слишком часто. Конечно, в силу Закона больших чисел вероятность такого события сходится к нулю, но у нас не так много данных.

Таким образом, вместо метки rand мы должны поставить либо 0, либо 1. Я на бумаге построил по этим правилам решающее дерево. Далее я заметил, что если вместо обоих меток rand написать 0, то дерево значительно упростится и приобретет вид:

--->

    3 класс  ---да---->  0

             ---нет--->  жен
         
                              ---да---->  1
                                        
                              ---нет--->  возр < 18
                              
                                                     ---да---->  1
                                                     
                                                     ---нет--->  0

Используя сложные решающие правила, полученное дерево можно заменить одним решающим правилом. А именно, выдаем 0, если

3 класс или (муж и возр > 17)

Посмотрим, насколько хорошо полученный классификатор классифицирует обучающую выборку. Ошибки при отнесению к классу 0 случаются в 23% случаях. То есть это случай, когда человек выживает, а классификатор говорит обратное. При отнесении к классу 1 ошибки случаются в 5% случаев. Это означает, что человек не выживает, а классификатор говорит обратное. В целом ошибки происходят в 21% случаев.
