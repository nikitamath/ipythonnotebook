```python
x = sample["HP"]
y = sample["MPG"]
a = mean( (x - mean(x)) * (y - mean(y)) ) / mean( (x - mean(x)) ** 2 )
b = mean(y) - a * mean(x)
a, b
```

Ошибка, посчитанная по формуле $err = \sqrt{ \frac {1} {n} \sum_{i = 1} ^ n (Y_i - (a X_i + b))^2 }$

```python
sqrt(sum((y - (a * x + b)) ** 2) / len(x))
```

### d).

Статистика $C_p$ имеет вид $\widehat{R} (S) = \widehat{R}_{tr} (S) + 2 |S| \widehat{\sigma}^2$, где $\widehat{R}_{tr} (S) = \sum_{i=1}^n (\widehat{Y}_i (S) - Y_i) ^ 2$ --- ошибка на обучающей выборке, $\widehat{\sigma} ^ 2 = \sum_{i=1}^n (X_i \widehat{\beta} - Y_i) ^ 2 / (n - k)$ --- подсчитано по всему множеству параметров.

Тут возникает вопрос, как подсчитывать значениеы этой статистики для признаков --- регрессию строить с константой или без. На слайде 20 написано, что в $S$ может не входить 1-ый признак, который константа. И это логично --- у нас набор признаков, мы хотим отобрать хорошие, может константа нам и не нужна. Но с другой стороны может по какому-то признаку все точки почти лежат на прямой $y = -x + 100$ равномерно по отрезку [0, 100]. Тогда с константой получится хорошее приближение по этому одному признаку, а без нее лучшее будет $y = x$ с огромной ошибкой. Я сделал обоими способами. Результаты получились разные.

##### Подсчет статистики без учета константы

Сначала для признака-константы

```python
y = sample["MPG"]
b = mean(y)
R_tr = sum((y - b) ** 2)
R = R_tr + 2 * sigma ** 2
R
```

Далее для всех остальных без константы. Они уже отсортированы.

```python
S = [[4], [1], [5], [2]]
for s in S:
    X = transpose(matrix(sample))[s]
    X = transpose(X)
    y = transpose(matrix(sample["MPG"]))
    beta = (((transpose(X) * X) ** (-1)) * transpose(X)) * y
    R_tr = sum((array(transpose(y - X * beta))[0] ** 2))
    R = R_tr + 2 * X.shape[1] * sigma ** 2
    print R, "\t", s
```

Теперь будет добавлять по одному признаку. Можно считать, что критерий изменился незначительно при добавлении признака в последний раз. Поэтому возьмем набор признаков (1, 4, 5, const) = (VOL, SP, WT, const).

```python
S = [[3], [3, 4], [3, 4, 1], [3, 4, 1, 5], [3, 4, 1, 5, 2]]
for s in S:
    X = transpose(matrix(sample))[s]
    X[0] = 1
    X = transpose(X)
    y = transpose(matrix(sample["MPG"]))
    beta = (((transpose(X) * X) ** (-1)) * transpose(X)) * y
    R_tr = sum((array(transpose(y - X * beta))[0] ** 2))
    R = R_tr + 2 * X.shape[1] * sigma ** 2
    print R, "\t", s[1:]
```

При обратном проходе получается такой же результат --- при выкидывании признака 5 (WT) значение увеличивается в 2 раза => стоит остановиться.

##### Подсчет статистики с учетом константы

Далее для всех остальных с константой. Они уже отсортированы. Можно заметить, что сдесь получился совсем другой порядок. Наименьшее значение статистики дает 5-ый признак (WT), хотя в предыдущем случае он без константы давал плохое значение.

```python
S = [[3,5], [3,2], [3,4], [3,1]]
for s in S:
    X = transpose(matrix(sample))[s]
    X[0] = 1
    X = transpose(X)
    y = transpose(matrix(sample["MPG"]))
    beta = (((transpose(X) * X) ** (-1)) * transpose(X)) * y
    R_tr = sum((array(transpose(y - X * beta))[0] ** 2))
    R = R_tr + 2 * X.shape[1] * sigma ** 2
    print R, "\t", s[1:]
```

Теперь будет добавлять по одному признаку. Тут явно видно, что критерий меняется незначительно при добавлении признака 2. Поэтому нужно взять набор признаков (5, const) = (WT, const). То есть один лишь 5-ый признак вместе с константой значим в этом подходе.

```python
S = [[3], [3, 5], [3, 5, 2], [3, 5, 2, 4], [3, 5, 2, 4, 1]]
for s in S:
    X = transpose(matrix(sample))[s]
    X[0] = 1
    X = transpose(X)
    y = transpose(matrix(sample["MPG"]))
    beta = (((transpose(X) * X) ** (-1)) * transpose(X)) * y
    R_tr = sum((array(transpose(y - X * beta))[0] ** 2))
    R = R_tr + 2 * X.shape[1] * sigma ** 2
    print R, "\t", s[1:]
```

При обратном подходе статистика сильно увеличится при удалении признака 4. Таким образом берем набор (2, 4, 5, const) = (HP, SP, WT, const).

Далее я решил исследовать, как признаки по отдельности вместе с константой подгоняют регрессию. Синим отмечены точки выборки. Кружки полупрозрачные, то есть при наложении цвет становится интенсивнее.

```python
for i in [1, 2, 4, 5]:
    X = transpose(matrix(sample))[[3, i]]
    X[0] = 1
    X = transpose(X)
    y = transpose(matrix(sample["MPG"]))
    beta = (((transpose(X) * X) ** (-1)) * transpose(X)) * y
    plt.figure(figsize=(17,4))
    plt.plot(X[:,1], y, 'bo', alpha=0.3)
    plt.plot(sort(X[:,1]), array(beta[1]) * array(sort(X[:,1])) + beta[0], 'r-')
    plt.show()
```

Отдельно стоит обратить внимание на последний график по пятому признаку (WT). В отличии от всех остальных случаев здесь почти все точки лежат совсем рядом с прямой. Только несколько точек сильно выбиваются. И можно заметить, что здесь зависимость имеет вид $y \sim -x + const$, то есть как раз тот случай, о котором я писал в самом начале пункта d).

Построим такие же графики, но без учета константы.

```python
for i in [1, 2, 4, 5]:
    X = transpose(matrix(sample))[[i]]
    X = transpose(X)
    y = transpose(matrix(sample["MPG"]))
    beta = (((transpose(X) * X) ** (-1)) * transpose(X)) * y
    plt.figure(figsize=(17,4))
    plt.plot(X, y, 'bo', alpha=0.3)
    plt.plot(sort(X), array(beta) * array(sort(X)), 'r-')
    plt.show()
```

Можно заметить, что на всех графиках подгонка очень плохая. Особенно замента разница с 5-м признаком.

### e).

Далее посмотрим для всевозможных комбинаций с константой

```python
S = [[3], [3,1], [3,2], [3,4], [3,5], [3,1,2], [3,1,4], [3,1,5], [3,2,4], [3,2,5], [3,4,5], [3,1,2,4],\
     [3,1,2,5], [3,1,4,5], [3,2,4,5], [3,1,2,4,5]]
for s in S:
    X = transpose(matrix(sample))[s]
    X[0] = 1
    X = transpose(X)
    y = transpose(matrix(sample["MPG"]))
    beta = (((transpose(X) * X) ** (-1)) * transpose(X)) * y
    R_tr = sum((array(transpose(y - X * beta))[0] ** 2))
    R = R_tr + 2 * X.shape[1] * sigma ** 2
    l = - len(y) * log(sigma) + (1 / (2 * sigma ** 2)) * sum((array(transpose(y - X * beta))[0] ** 2))
    BIC = -l - X.shape[1] * log(len(y)) / 2
    print R, "\t", BIC, "\t", s[1:]
```

И без константы

```python
S = [[1], [2], [4], [5], [1,2], [1,4], [1,5], [2,4], [2,5], [4,5], [1,2,4],\
     [1,2,5], [1,4,5], [2,4,5], [1,2,4,5]]
for s in S:
    X = transpose(matrix(sample))[s]
    X = transpose(X)
    y = transpose(matrix(sample["MPG"]))
    beta = (((transpose(X) * X) ** (-1)) * transpose(X)) * y
    R_tr = sum((array(transpose(y - X * beta))[0] ** 2))
    R = R_tr + 2 * X.shape[1] * sigma ** 2
    l = - len(y) * log(sigma) + (1 / (2 * sigma ** 2)) * sum((array(transpose(y - X * beta))[0] ** 2))
    BIC = -l - X.shape[1] * log(len(y)) / 2
    print R, "\t", BIC, "\t", s
```

Видим, что статистика $C_p$ принимает наименьшее значение на наборе признаков (2, 4, 5, const) = (HP, SP, WT, const), что совпадает с результатом обратного прохода при сортировке с учетом константы. Этот же результат дает статистика BIC.
