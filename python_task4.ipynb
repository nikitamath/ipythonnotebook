```python
def print_grafic(lists, colors, description, x_label, y_label, figsize=(18, 4)):
    plt.figure(figsize=figsize)
    for l, c, d in zip(lists, colors, description):
        plt.plot(range(len(l)), l, c, label=d)
    legend = plt.legend(loc='upper right', shadow=True, fontsize='x-large')
    plt.xlabel(x_label, fontsize=12)
    plt.ylabel(y_label, fontsize=12)
    plt.show()
```

Ниже приведены графики для радиусов обзора от 4 до 1 в цвета зеленый, синий, желтый и красный соответственно. Видно, что начиная с некоторого периода графики не сильно различаются. Для 0-го радиуса обзора (случайные действия хищников), все хищники быстро вымирают.

```python
print_grafic([predator_count_4, predator_count_3, predator_count_2, predator_count_1],
             ['g', 'b', 'y', 'r'], ['4', '3', '2', '1'], 'iterations', 'predator count')
print_grafic([food_count_4, food_count_3, food_count_2, food_count_1],
             ['g', 'b', 'y', 'r'], ['4', '3', '2', '1'], 'iterations', 'food count')
```
