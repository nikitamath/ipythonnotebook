### 1. Определение функции

```python
from scipy.stats import norm
F_inf = norm(1, 1)
F_0 = norm(0, 1)
```

```python
def T_cusum(B, filename, F_true, F_alt, F_gen, boot_size = 2000):
    l = len(B)
    T = zeros(l)
    for j in range(l):
        tau = zeros(boot_size) - 1
        for i in range(boot_size):
            gamma = 0
            while gamma < B[j]:
                X = F_gen.rvs(1)
                gamma = max(0, gamma + (-X + 0.5))#log (F_alt.pdf(X) / F_true.pdf(X)))
                tau[i] += 1

        T[j] = mean(tau)
    
    f = open(filename, 'w')
    f.write('boot_size=' + str(boot_size))
    f.write('\n')
    for j in range(l):
        f.write(str(B[j]))
        f.write(',')
        f.write(str(T[j]))
        f.write('\n')
    f.close()
    return T
```

Здесь закомментированна часть строки, а вместо нее написано "(-X + 0.5))". Дело в том, что явно подсчитанный логарифм отношения правдоподобия ускоряет вычисления в несколько раз. Фактически, теперь 3 и 4 аргументы у функции не нужны.

### 3. Анализ зависимости $T(B)$

Для начала построим график полученной зависимости. Отдельно также график для малых значений $B$, чтобы было лучше видно.

```python
plt.figure(figsize=(17,6))
plt.plot(BT['B'], BT['T'], 'bo-')
plt.show()
plt.figure(figsize=(17,6))
plt.plot(BT['B'][0:45], BT['T'][0:45], 'bo-')
plt.show()
```

Оценим зависисмость с помощью МНК. Из графика видно, что зависимость напиминает экпоненту.
$$T = \beta_0 + \beta_1 e^B + \varepsilon$$
МНК-оценка:
$$\widehat{\beta}_1 = \frac {\sum_i{(T_i - \overline{T}) (X_i - \overline{X})}} {\sum_i{(X_i - \overline{X})^2}}, X_i = e^{B_i}$$
$$\widehat{\beta}_0 = \overline{T} - \widehat{\beta}_1 \overline{X}$$

```python
X = e ** array(BT['B'])
T = array(BT['T'])
beta_1 = sum((T - mean(T)) * (X - mean(X))) / sum((X - mean(X)) ** 2)
beta_0 = mean(T) - beta_1 * mean(X)
print beta_0, beta_1
```

Построим графики

```python
B = (1 + array(range(110))) / 20.
plt.figure(figsize=(17,4))
plt.plot(B, beta_1 * e ** B + beta_0, 'g-')
plt.plot(BT['B'], BT['T'], 'b*')
plt.show()
B = (1 + array(range(50))) / 25.
plt.figure(figsize=(17,4))
plt.plot(B, beta_1 * e ** B + beta_0, 'g-')
plt.plot(BT['B'][0:45], BT['T'][0:45], 'b*')
plt.show()
```

Построим также график остатков

```python
plt.figure(figsize=(17,4))
plt.plot(BT['B'], abs(beta_1 * e ** BT['B'] + beta_0 - BT['T']), 'ro-')
plt.show()
```

Среднеквадратичная ошибка

```python
mean((beta_1 * e ** BT['B'] + beta_0 - BT['T']) ** 2)
```

Видно, что экспонента настолько хорошо ложится, что даже отпадает желание строить регрессию для других подобных функций, например, для квадрата.
Проанализируем более детально полученные результаты. Видим, что при малых значениях $B$ (меньших 0.5) ошибка получается большой по сравнению с принимаемыми значениями. При больших значениях $B$ ошибка не более 8 по сравнению с принимаемыми значениями (порядка 1000) становится несущественной. Посмотрим более внимательно на значения при малых $B$.

```python
plt.figure(figsize=(17,6))
plt.plot(BT['B'][0:30], BT['T'][0:30], 'bo-')
plt.show()
```

Видим, что при $B < 0.5$ зависимость сильно напоминает линейную. Давайте тогда при $B < t$ приближать линейной зависимостью, а при $B > t$ --- экпоненциальной. Параметр $t$ подберем по критерию $R^2$, который эквивалентен минимизации среднеквадратичной ошибки.

```python
TTT = range(20, 40)
err = array(zeros(20))
for t in TTT:
    X_lin = array(BT['B'])[:t]
    X_exp = e ** array(BT['B'])[t:]
    T = array(BT['T'])
    beta_exp_1 = sum((T[t:] - mean(T[t:])) * (X_exp - mean(X_exp))) / sum((X_exp - mean(X_exp)) ** 2)
    beta_exp_0 = mean(T[t:]) - beta_exp_1 * mean(X_exp)
    beta_lin_1 = sum((T[:t] - mean(T[:t])) * (X_lin - mean(X_lin))) / sum((X_lin - mean(X_lin)) ** 2)
    beta_lin_0 = mean(T[:t]) - beta_lin_1 * mean(X_lin)
    err[t - 20] = 1 - ((sum((beta_exp_1 * e ** BT['B'][t:] + beta_exp_0 - BT['T'][t:]) ** 2) + \
        sum((beta_lin_1 * BT['B'][:t] + beta_lin_0 - BT['T'][:t]) ** 2)) / len(BT['B'])) / var(T)

plt.figure(figsize=(17,4))
plt.plot(TTT, err, 'ro-')
plt.show()
```

Максимум достигается на 31-й точке, которая соответствует $t = 1.1$. Таким образом, при $B < 1.1$ будем приближать линейной функцией, а при $B > 1.1$ --- экспоненциальной.

```python
t = 31
X_lin = array(BT['B'])[:t]
X_exp = e ** array(BT['B'])[t:]
T = array(BT['T'])
beta_exp_1 = sum((T[t:] - mean(T[t:])) * (X_exp - mean(X_exp))) / sum((X_exp - mean(X_exp)) ** 2)
beta_exp_0 = mean(T[t:]) - beta_exp_1 * mean(X_exp)
beta_lin_1 = sum((T[:t] - mean(T[:t])) * (X_lin - mean(X_lin))) / sum((X_lin - mean(X_lin)) ** 2)
beta_lin_0 = mean(T[:t]) - beta_lin_1 * mean(X_lin)
print 'lin: ', beta_lin_0, beta_lin_1
print 'exp: ', beta_exp_0, beta_exp_1
```

### Численное решение интегральных уравнений

Определение функции, подсчет и запись в файл, а потом чтение из файла.

```python
def solve_eq(B, l_B, F):
    x = array([i * B * 1.0 / l_B for i in range(l_B + 1)])
    A = transpose(matrix(zeros(l_B + 1))) * matrix(zeros(l_B + 1))
    A[0] = 1 - F.cdf(x + 0.5)
    for i in range(1, l_B + 1):
        A[i] = F.pdf(x[i] + 0.5 - x) * B / l_B
    A = transpose(A)
    g = - ((A - eye(l_B + 1)) ** (-1)) * transpose((1 + matrix(zeros(l_B + 1))))
    return g[0]
```

```python
l_B = 500
B = (1 + array(range(100))) / 20.
f = open('eqT_1.txt', 'w')
f.write('l_B=' + str(l_B))
f.write('\n')
for b in B:
    f.write(str(b))
    f.write(',')
    g = array(solve_eq(b, l_B, F_inf))[0][0]
    f.write(str(g))
    f.write('\n')

f.close()
```

```python
l_B = 500
B = (1 + array(range(200))) / 20.
f = open('eqR_1.txt', 'w')
f.write('l_B=' + str(l_B))
f.write('\n')
for b in B:
    f.write(str(b))
    f.write(',')
    g = array(solve_eq(b, l_B, F_0))[0][0]
    f.write(str(g))
    f.write('\n')

f.close()
```

График для зависисмости $T(B)$. Синим --- значения по методу Монте-Карло, красным --- значения из численного решения уравнения. До $B = 0.5$ значения практически совпадают. Дальше значения из численного решения уравнения начинают скакать и даже уходят в отрицательную область. Это связано с тем, что при численном решении матрица получается близкая к вырожденной.
