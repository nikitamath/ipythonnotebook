### Графики зависимости статистики $Q$ от времени

Обучаем модель. Я беру 20 гравных компонент

```python
d_optim = 20
pca = PCA(n_components=d_optim)
pca.fit(X2)
```

Первая неделя

```python
Q_X2 = []
for x in X2:
    error = matrix(x - pca.inverse_transform(pca.transform(x)))
    Q_X2.append((error * transpose(error))[0,0])
plt.figure(figsize=(18,6))
plt.plot(range(len(Q_X2)), Q_X2, 'b-')
plt.xlim((0, len(Q_X2)))
plt.show()
```

Во второй недели получился один большой выброс и один маленький

```python
Q_X3 = []
for x in X3:
    error = matrix(x - pca.inverse_transform(pca.transform(x)))
    Q_X3.append((error * transpose(error))[0,0])
plt.figure(figsize=(18,6))
plt.plot(range(len(Q_X3)), Q_X3, 'b-')
plt.xlim((0, len(Q_X3)))
plt.show()
```

Чтобы лучше разглядеть график, ограничим область видимых значений по игрику

```python
plt.figure(figsize=(18,6))
plt.plot(range(len(Q_X3)), Q_X3, 'b-')
plt.xlim((0, len(Q_X3)))
plt.ylim((0, 10 * Q_X3[0]))
plt.show()
```

Третья неделя

```python
Q_X4 = []
for x in X4:
    error = matrix(x - pca.inverse_transform(pca.transform(x)))
    Q_X4.append((error * transpose(error))[0,0])
plt.figure(figsize=(18,6))
plt.plot(range(len(Q_X4)), Q_X4, 'b-')
plt.xlim((0, len(Q_X4)))
plt.show()
```

### Порог $Q_{\alpha}$

```python
from scipy.stats import norm
```

Вычислим значение порога по данным первой недели

```python
alpha_1 = 0.05
c_1 = norm.ppf(alpha_1)
alpha_2 = 0.01
c_2 = norm.ppf(alpha_2)

lambdas_X2 = array(lambdas_X2)
theta_1 = sum(lambdas_X2[:d_optim])
theta_2 = sum(lambdas_X2[:d_optim] ** 2)
theta_3 = sum(lambdas_X2[:d_optim] ** 3)
h_0 = 1 - 2 * theta_1 * theta_3 / (3 * theta_2 ** 2)
Q_alpha1 = theta_1 * ((c_1 * sqrt(2 * theta_2 * h_0 ** 2)) / theta_1 + \
                        (theta_2 * h_0 * (h_0 - 1)) / (theta_1 ** 2) + 1) ** (1 / h_0)
Q_alpha2 = theta_1 * ((c_2 * sqrt(2 * theta_2 * h_0 ** 2)) / theta_1 + \
                        (theta_2 * h_0 * (h_0 - 1)) / (theta_1 ** 2) + 1) ** (1 / h_0)

print Q_alpha1, Q_alpha2
```

Для первой недели один из порогов вообще не достигается, а второй лишь слегка достигается. Это связано с тем, что мы подгоняем модель под данные первой недели.

```python
plt.figure(figsize=(18,6))
plt.plot(range(len(Q_X2)), Q_X2, 'b-')
plt.plot(range(len(Q_X2)), Q_alpha1 + zeros(len(Q_X2)), 'r-')
plt.plot(range(len(Q_X2)), Q_alpha2 + zeros(len(Q_X2)), 'r-')
plt.xlim((0, len(Q_X2)))
plt.show()
```

На второй недели для нижнего порога есть даже непрерывный участок, где он достигается

```python
plt.figure(figsize=(18,6))
plt.plot(range(len(Q_X3)), Q_X3, 'b-')
plt.plot(range(len(Q_X3)), Q_alpha1 + zeros(len(Q_X3)), 'r-')
plt.plot(range(len(Q_X3)), Q_alpha2 + zeros(len(Q_X3)), 'r-')
plt.xlim((0, len(Q_X3)))
plt.ylim((0, 10 * Q_X3[0]))
plt.show()
```

На третьей недели так же верхний порог почти не достигается

```python
plt.figure(figsize=(18,6))
plt.plot(range(len(Q_X4)), Q_X4, 'b-')
plt.plot(range(len(Q_X4)), Q_alpha1 + zeros(len(Q_X4)), 'r-')
plt.plot(range(len(Q_X4)), Q_alpha2 + zeros(len(Q_X4)), 'r-')
plt.xlim((0, len(Q_X4)))
plt.show()
```

# 2. Детектирование аномалий на основе One-Class SVM

### Чтение и техническое преобразование данных

```python
train = np.loadtxt('kddcup.data_10_percent', delimiter=',', dtype=str)
test = np.loadtxt('KDD_test_v2.csv', delimiter=';', dtype=str)
```

```python
print train.shape, test.shape
```

```python
train = transpose(train)
test = transpose(test)
```

```python
feature1 = dict(zip(train[1], range(len(train[1]))))
feature2 = dict(zip(train[2], range(len(train[2]))))
feature3 = dict(zip(train[3], range(len(train[3]))))
for key, value in zip(feature1.keys(), range(len(feature1.keys()))):
    feature1[key] = value + 1
for key, value in zip(feature2.keys(), range(len(feature2.keys()))):
    feature2[key] = value + 1
for key, value in zip(feature3.keys(), range(len(feature3.keys()))):
    feature3[key] = value + 1
```

```python
y_train = array([2 * int(y == 'normal.') - 1 for y in train[41]])
```

```python
train_num = [transpose(map(float, train[i])) for i in [0] + range(4, 41)]
train_num += [[feature1[label] for label in train[1]]]
train_num += [[feature2[label] for label in train[2]]]
train_num += [[feature3[label] for label in train[3]]]
train_num = transpose(array(train_num))
```

```python
def func(key, dictionary):
    if key in dictionary:
        return dictionary[key]
    else:
        return 0
```

```python
test_num = [transpose(map(float, test[i])) for i in [0] + range(4, 41)]
test_num += [[func(label, feature1) for label in test[1]]]
test_num += [[func(label, feature2) for label in test[2]]]
test_num += [[func(label, feature3) for label in test[3]]]
test_num = transpose(array(test_num))
```

```python
print array(zip(train_num, y_train))[0]
```

```python
train_normal = transpose(filter(lambda x: 1 + x[1], zip(train_num, y_train)))[0]
train_normal = array(map(list, train_normal))
print train_normal.shape
```

```python
print train_num
print test_num
print y_train
```

### Обучение One-class SVM

```python
from sklearn import svm
```

```python
import time
import random
```

Так предполагалось запускать алгоритм обучения

```python
clf = svm.OneClassSVM(nu=0.05, kernel="rbf", gamma=0.1, max_iter=500)
clf.fit(train_normal)
```

Но работает очень долго, и для полной кросс-проверки потребуется полгода. Поэтому я решил обучаться и тестироваться на небольших подвыборках. Здесь написана функция, которая генерирует случайные подвыборки, обучает на них модель и считает ошибку.

```python
def iterations(size_train, size_test, num_iter, nu=0.1, kernel="rbf", gamma=0.2):
    time_training = []
    time_predicting = []
    time_generating = []
    errors = []

    for i in range(num_iter):
        begin = time.clock()
        order_train = range(n_normal)
        random.shuffle(order_train)
        indexs_train = order_train[:size_train]
        X_train_part = train_normal[indexs_train]

        order_test = range(n)
        random.shuffle(order_test)
        indexs_test = order_test[:size_test]
        X_test_part = train_num[indexs_test]
        y_test_part = y_train[indexs_test]
        time_generating.append(time.clock() - begin)

        clf = svm.OneClassSVM(nu=nu, kernel=kernel, gamma=gamma, max_iter=100)

        begin = time.clock()
        try:
            clf.fit(X_train_part)
        except Exception:
            print i, Exception
            continue
        time_training.append(time.clock() - begin)

        begin = time.clock()
        predict = clf.predict(X_test_part)
        time_predicting.append(time.clock() - begin)

        errors.append(sum(abs(y_test_part - predict)) / (2 * size_test))
    
    return mean(errors), mean(time_training), mean(time_predicting), mean(time_generating)
```

Параметры тестирования

```python
n_normal = len(train_normal)
size_train = n_normal / 100

n = len(train_num)
size_test = n / 100

num_iter = 100
```

#### Исследование зависимости ошибки от $\nu$

```python
grid_size = 20
nu = np.linspace(0.001, 0.5, grid_size)
nu_results = []
for i in nu:
    print i
    nu_results.append(iterations(size_train, size_test, num_iter, nu=i))
```

```python
grid_size = 20
nu1 = np.linspace(0.5, 1, grid_size)
nu1_results = []
for i in nu1:
    print i
    nu1_results.append(iterations(size_train, size_test, num_iter, nu=i))
```
