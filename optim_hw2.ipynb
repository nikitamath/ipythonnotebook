```python
import numpy as np
import matplotlib.pyplot as plt
from time import time
import scipy.stats as sps
import copy
import math
```

## 1. Безусловная минимизация невыпуклой функции

$$f(x) = \frac14 (x_1 - 1)^2 + \sum\limits_{i = 2}^n (2 x_{i-1}^2 - x_i - 1)^2$$

Очевидно, что функция неотрицательна и равна нулю в точке $(1, ..., 1)$.

### Производные

#### Первые производные

$$\frac {\partial f} {\partial x_1} = \frac {x_1 - 1} {2} + 8 x_1 (2 x_1^2 - x_2 - 1)$$

$$\frac {\partial f} {\partial x_i} = 8 x_i (2 x_i^2 - x_{i+1} - 1) - 2(2 x_{i-1}^2 - x_i - 1),\ \ \ \ i = 2, ..., n - 1$$

$$\frac {\partial f} {\partial x_n} = - 2(2 x_{n-1}^2 - x_n - 1)$$

#### Вторые производные

$$\frac {\partial^2 f} {\partial x_1^2} = 48 x_1^2 - 8 x_2 - 7.5$$

$$\frac {\partial^2 f} {\partial x_1 \partial x_2} = -8 x_1$$

$$\frac {\partial^2 f} {\partial x_i^2} = 48 x_i^2 - 8 x_{i+1} - 6,\ \ \ \ i = 2, ..., n - 1$$
$$\frac {\partial^2 f} {\partial x_i \partial x_{i+1}} = -8 x_i,\ \ \ \ i = 2, ..., n - 1$$

$$\frac {\partial^2 f} {\partial x_n^2} = 2$$

$$f'' (x) = 
\begin{pmatrix} 
        48 x_1^2 - 8 x_2 - 7.5 & -8 x_1 & 0 & ... & 0 & 0\\
 	 -8 x_1 & 48 x_2^2 - 8 x_3 - 6 & -8 x_2 & ... & 0 & 0\\
	 0 & -8 x_2 & 48 x_3^2 - 8 x_4 - 6 & ... & 0 & 0\\
        ... &  &  &  & & \\
        0 & 0 & 0 & ... & 48 x_{n-1}^2 - 8 x_n - 6 & -8 x_{n-1} \\
        0 & 0 & 0 & ... & -8 x_{n-1} & 2 \\
\end{pmatrix}.$$


####Случай $n = 2$

$$f(x) = \frac14 (x_1 - 1)^2 + (2 x_1^2 - x_2 - 1)^2$$

$$f' (x) = 
\begin{pmatrix} 
        \frac {x_1 - 1} {2} + 8 x_1 (2 x_1^2 - x_2 - 1)\\
 	 - 2(2 x_1^2 - x_2 - 1)\\
\end{pmatrix}.$$
$$f' (x) = \left( \frac {x_1 - 1} {2} + 8 x_1 (2 x_1^2 - x_2 - 1), - 2(2 x_1^2 - x_2 - 1) \right)$$

$$f'' (x) = 
\begin{pmatrix} 
        48 x_1^2 - 8 x_2 - 7.5\ & -8 x_1\\
 	 -8 x_1\ & 2\\
\end{pmatrix}.$$


### Методы

##### 1. Градиентный метод

$$x^{k+1} = x^k - \gamma^k f' (x^k)$$

##### 2. Метод наискорейшего спуска

$$x^{k+1} = x^k - \gamma^k f' (x^k),$$
где $\gamma^k = \mathop{\rm arg\,min}\limits_{\gamma \geqslant 0} f( x^k - \gamma f' (x^k) )$.

##### 3. Метод покоординатного спуска спуска

$$x^{k+1} = x^k - \gamma^k \frac {\partial f} {\partial x_i} (x^k)$$

##### 4. Метод тяжелого шара

$$x^{k+1} = x^k - \gamma^k f' (x^k) - \beta (x^k - x^{k+1})$$

##### 5. Метод сопряженных градиентов

$$x^{k+1} = x^k + \alpha p^k,$$
где $p^k = -f' (x^k) + \beta^k p^{k-1}$, $\alpha = \mathop{\rm arg\,min}\limits_{\alpha} f( x^k + \alpha p^k )$, $\beta^k = \frac {||f' (x^k)||^2} {||f' (x^{k-1})||}$, $\beta_0 = 0$.

##### 6. Метод Ньютона

$$x^{k+1} = x^k - \left( f'' (x^k) \right)^{-1} f' (x^k)$$

### Функция, градиент и гессиан

```python
def f(x):
    value = (x[0] - 1) ** 2 / 4.
    for i in range(1, len(x)):
        value += (2 * x[i - 1] ** 2 - x[i] - 1) ** 2
    return value

def grad_f(x, axis=None):
    def calculate_for_axis(axis):
        if axis == 0:
            return (x[0] - 1) / 2. + 8 * x[0] * (2 * x[0] ** 2 - x[1] - 1)
        elif axis == len(x) - 1:
            return -2 * (2 * x[-2] ** 2 - x[-1] - 1)
        else:
            return 8 * x[axis] * (2 * x[axis] ** 2 - x[axis + 1] - 1) - 2 * (2 * x[axis - 1] ** 2 - x[axis] - 1)
    
    if axis is not None:
        return calculate_for_axis(axis)
    else:
        values = np.matrix(np.zeros((len(x), 1)))
        for axis in range(len(x)):
            values[axis] = calculate_for_axis(axis)
    return values

def hessian_f(x):
    hessian = np.matrix(np.zeros((len(x), len(x))))
    hessian[0, 0] = 48 * x[0] ** 2 - 8 * x[1] - 7.5
    hessian[0, 1] = -8 * x[0]
    for i in range(1, len(x) - 1):
        hessian[i, i - 1] = -8 * x[i - 1]
        hessian[i, i] = 48 * x[i] ** 2 - 8 * x[i + 1] - 6
        hessian[i, i + 1] = -8 * x[i]
    hessian[-1, -2] = -8 * x[-2]
    hessian[-1, -1] = 2
    return hessian
```

### Реализация шага методов

```python
def gradient_method_step(x, grad_f, gamma=0.1):
    return x - gamma * grad_f(x)

def steepest_descent_method_step(x, grad_f, step_max=0.3, grid_step=0.01):
    grid = np.arange(0, step_max, grid_step)
    f_grid = [f(x - gamma * grad_f(x)) for gamma in grid]
    return x - grid[np.argmin(f_grid)] * grad_f(x)

def stochastic_gradient_descent_method_step(x, grad_f, gamma=0.1):
    x = copy.deepcopy(x)
    axis = sps.randint.rvs(size=1, low=0, high=len(x))
    x[axis] = x[axis] - gamma * grad_f(x, axis=axis)
    return x

def heavy_ball_method_step(x, grad_f, gamma=0.1, beta=0.1):
    x_prev, x_cur = x
    return x_cur, x_cur - gamma * grad_f(x_cur) - beta * (x_cur - x_prev)

def conjugate_gradient_method(x, grad_f, step_max=0.1, grid_step=0.001):
    x_prev, p_prev, x_cur = x
    beta = np.sum(np.array(grad_f(x_cur)) ** 2) / sqrt(np.sum(np.array(grad_f(x_prev)) ** 2))
    p_cur = -grad_f(x_cur) + beta * p_prev
    
    alpha_grid = np.arange(0, step_max, grid_step)
    f_grid = [f(x_cur + alpha * p_cur) for alpha in alpha_grid]

    return x_cur, p_cur, x_cur + alpha_grid[np.argmin(f_grid)] * p_cur

def newton_method_step(x, grad_f, hessian_f):
    return x - hessian_f(x) ** (-1) * grad_f(x)
```

Функция, которая для данного метода рисует последовательнось итераций

```python
def graw_method_steps(method, start_point=np.matrix([[0], [0]]), n_iter=10, figsize=(7, 7), title=''):
    grad_x = [start_point]
    for i in range(n_iter):
        grad_x.append(method(grad_x[-1]))

    x1_values = [float(p[0]) for p in grad_x]
    x2_values = [float(p[1]) for p in grad_x]

    X, Y = np.mgrid[min(x1_values) - 0.5 : max(x1_values) + 0.5 : 0.01, min(x2_values) - 0.5 : max(x2_values) + 0.5 : 0.01]
    Z = [f((x, y)) for (x, y) in zip(X, Y)]

    plt.figure(figsize=figsize)
    CS = plt.contour(X, Y, Z, [float(f(p)) for p in grad_x])
    plt.clabel(CS, fontsize=8, inline=1)
    plt.plot(x1_values, x2_values, 'o-', color='black')
    plt.title(title)
    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.show()
```

### Оптимизация в 2D

###### Градиентный метод
Здесь видно, что $\gamma = 0.053$ --- примерно оптимальный шаг, и за $7$ итераций метод уже близко подходит к минимуму.
Дело в том, что у него сразу получается перепрыгнуть близкую область малых значений и приблизиться к минимуму.
В тоже время, если взять маленький $\gamma = 0.01$, то метод сходится очень медленно --- не успевает перепрыгнуть область малых значений и очень медленно по ней передвигается.
Если взять большое (но не слишком) $\gamma = 0.09$, то его сначала пошатает, а потом в итоге опять попадет в плохую область малых значений и будет долго сходиться.
Если взять его чуть больше --- $\gamma = 0.095$, то он быстро улетит в бесконечность.

```python
for gamma, n_iter in zip([0.053, 0.01, 0.09, 0.095], [7, 1000, 100, 7]):
    graw_method_steps(lambda point: gradient_method_step(point, grad_f, gamma=gamma), 
                      np.matrix([[-1.5], [1]]), 
                      n_iter=n_iter,
                      title='gradient method with gamma={}, n_iter={}'.format(gamma, n_iter))
```

###### Метод наискорейшего спуска
Быстро сходится в близкую к минимуму область, но потом долго спускается по ней.

```python
n_iter = 200
graw_method_steps(lambda point: steepest_descent_method_step(point, grad_f, grid_step=0.001), 
                  np.matrix([[-1.5], [1]]), 
                  n_iter=n_iter,
                  title='steepest descent method with n_iter={}'.format(n_iter))
```

###### Метод стохастического градиента
Достаточно быстро сходится в близкую к минимуму область, и потом так же долго спускается по ней.

```python
n_iter = 200
graw_method_steps(lambda point: stochastic_gradient_descent_method_step(point, grad_f, gamma=0.1), 
                  np.matrix([[-1.5], [1]]), 
                  n_iter=n_iter,
                  title='stochastic gradient descent method with n_iter={}'.format(n_iter))
```

###### Метод тяжелого шара
При большом $\beta$ тяжесть ему только мешает.
Зато картинка получилась красивой :)

```python
for beta, n_iter in zip([0.12, 0], [50, 30]):
    def step_func(point):
        prev, x_new = heavy_ball_method_step([point.prev, point], grad_f, gamma=0.05, beta=beta)
        x_new.prev = prev
        return x_new

    x0 = np.matrix([[-1.5], [1]])
    x0.prev = np.matrix([[-1.5], [1]])
    graw_method_steps(step_func, 
                      x0, 
                      n_iter=n_iter,
                      title='heavy ball method with beta = {}, n_iter={}'.format(beta, n_iter))
```

###### Метод сопряженных градиентов
Достаточно быстро сходится в близкую к минимуму область, и потом долго спускается по ней.

```python
for n_iter in [300]:
    def step_func(point):
        x_prev, p_prev, x_new = conjugate_gradient_method([point.x_prev, point.p_prev, point], grad_f)
        x_new.x_prev = x_prev
        x_new.p_prev = p_prev
        return x_new

    x0 = np.matrix([[-1.5], [1]])
    x0.x_prev = np.matrix([[-1.5], [1]])
    x0.p_prev = np.matrix([[0], [0]])
    graw_method_steps(step_func, 
                      x0, 
                      n_iter=n_iter,
                      title='conjugate gradient method with n_iter={}'.format(n_iter))
```

###### Метод Ньютона
Он оказался чемпионом --- всего 5 итераций.

```python
n_iter = 5
graw_method_steps(lambda point: newton_method_step(point, grad_f, hessian_f), 
                  np.matrix([[-1.5], [1]]), 
                  n_iter=n_iter,
                  title='newton method with n_iter={}'.format(n_iter))
```

### Общий случай

Функции генерации начальной точки и запуска итераций метода с критериями остановки.

```python
def get_start_point(dim):
    point = np.ones(dim)
    point[0] = -1.5
    return np.matrix(point).T

def run_iterations(method, grad_f, start_point=get_start_point(2), n_iter=10, time_limit=None, accuracy=None):
    points = [start_point]
    iteration_number = 0
    t = time()
    while True:
        norm_grad = np.linalg.norm(grad_f(points[-1]))
        if (n_iter is not None and iteration_number >= n_iter) \
           or (time_limit is not None and time() - t >= time_limit) \
           or (accuracy is not None and norm_grad <= accuracy) \
           or math.isnan(norm_grad) \
           or norm_grad > 10 ** 7:
                break
        
        points.append(method(points[-1]))
    
    return points, time() - t
```

Функция автоматической генерации html-таблицы для вставки в ноутбук.

```python
def create_table_line(points, times, descriptions, grad_f, accuracy, xmin=np.matrix(np.ones(len(points[0][-1]))).T, fmin=0):
    table = '<table>\n<tr>\n\t<td>Метод</td>\n\t<td>$n$</td>\n\t<td>$N$</td>' + \
            '\n\t<td width=\"20%\">$||\\nabla f(\widehat{x})||$</td>' + \
            '\n\t<td>$||x^* - \widehat{x}||$</td>\n\t<td>$|f^* - f(\widehat{x})|$</td>' + \
            '\n\t<td>t</td>\n\t<td>Сошелся?</td>\n</tr>\n'
            
    for p, t, d in zip(points, times, descriptions):
        table += '<tr>\n'
        table += '\t<td>{}</td>\n'.format(d)
        table += '\t<td>{}</td>\n'.format(len(p[0]))
        table += '\t<td>{}</td>\n'.format(len(p) - 1)
        try:
            table += '\t<td>{}</td>\n'.format(sqrt(np.sum(np.array(grad_f(p[-1])) ** 2)))
        except OverflowError:
            table += '\t<td>Overflow</td>\n'
        
        xmin=np.matrix(np.ones(len(p[-1]))).T
        table += '\t<td>{}</td>\n'.format(sqrt(np.sum(np.array(p[-1] - xmin) ** 2)))
        table += '\t<td>{}</td>\n'.format(sqrt(np.sum(np.array(f(p[-1]) - fmin) ** 2)))
        table += '\t<td>{}</td>\n'.format(round(t, 1))
        try:
            table += '\t<td>{}</td>\n'.format(np.linalg.norm(grad_f(p[-1])) <= accuracy)
        except OverflowError:
            table += '\t<td>False</td>\n'
        table += '</tr>\n'
        
    
    table += '</table>'
    return table
```

Запуск всех методов

```python
points, times, descriptions = [], [], []
time_limit = 5 * 60
accuracy=10 ** (-8)

for n, gamma in zip([2, 2, 3, 3, 4], [0.05, 0.03, 0.03, 0.01, 0.03]):
    p, t = run_iterations(lambda point: gradient_method_step(point, grad_f, gamma=gamma), 
                          grad_f, start_point=get_start_point(n), n_iter=None, time_limit=time_limit, accuracy=accuracy)
    points.append(p), times.append(t), descriptions.append('Градиентный метод с $\gamma$={}'.format(gamma))

for n in zip([2, 3, 4]):
    p, t = run_iterations(lambda point: steepest_descent_method_step(point, grad_f), 
                          grad_f, start_point=get_start_point(n), n_iter=None, time_limit=time_limit, accuracy=accuracy)
    points.append(p), times.append(t), descriptions.append('Метод наискорейшего спуска')
    
for n in zip([2, 3, 4]):
    p, t = run_iterations(lambda point: stochastic_gradient_descent_method_step(point, grad_f, gamma=0.05), 
                          grad_f, start_point=get_start_point(n), n_iter=None, time_limit=time_limit, accuracy=accuracy)
    points.append(p), times.append(t), descriptions.append('Метод стохастического градиента')
    
for n in zip([2, 3, 4]):
    def step_func(point):
        prev, x_new = heavy_ball_method_step([point.prev, point], grad_f, gamma=0.05, beta=0.1)
        x_new.prev = prev
        return x_new
    
    x0 = get_start_point(n)
    x0.prev = get_start_point(n)
    p, t = run_iterations(step_func, grad_f, start_point=x0,
                          n_iter=None, time_limit=time_limit, accuracy=accuracy)
    points.append(p), times.append(t), descriptions.append('Метод тяжелого шара')
    
for n in zip([2, 3, 4]):
    def step_func(point):
        x_prev, p_prev, x_new = conjugate_gradient_method([point.x_prev, point.p_prev, point], grad_f)
        x_new.x_prev = x_prev
        x_new.p_prev = p_prev
        return x_new

    x0 = get_start_point(n)
    x0.x_prev = get_start_point(n)
    x0.p_prev = get_start_point(n) * 0
    
    p, t = run_iterations(step_func, grad_f, start_point=x0, 
                          n_iter=None, time_limit=time_limit, accuracy=accuracy)
    points.append(p), times.append(t), descriptions.append('Метод сопряженных градиентов')
    
for n in zip([2, 3, 4, 5, 6, 7]):
    p, t = run_iterations(lambda point: newton_method_step(point, grad_f, hessian_f), 
                          grad_f, start_point=get_start_point(n), n_iter=None, time_limit=time_limit, accuracy=accuracy)
    points.append(p), times.append(t), descriptions.append('Метод Ньютона')
```

## 2. Безусловная минимизация выпуклой функции

$$f(x) = \sum\limits_{i=0}^n e^{-x_i} + e^{\sum\limits_{i=0}^n x_i - 1}$$

$$\frac {\partial f} {\partial x_i} = -e^{-x_i} + e^{\sum\limits_{i=0}^n x_i - 1}$$

$$\frac {\partial^2 f} {\partial x_i \partial x_j} = e^{-x_i} I\{i = j\} + e^{\sum\limits_{i=0}^n x_i - 1}$$

Из равенства нулю производных получаем, что минимум достигается в точке $x^* = \frac {1} {n + 1} (1, ..., 1)$ и значение минимума равно $f^* = (n + 1) e^{-\frac {1} {n + 1}}$.

```python
V, _ = np.linalg.qr(sps.uniform.rvs(size=(100, 100)))
D = np.diag(np.ones(100))
D[0, 0] = 10 ** 6
P = V * D * V.T
```

```python
def f2(x):
    return np.sum(e ** (-np.array(x))) + e ** (np.sum(x) - 1)

def grad_f2(x, axis=None):
    exp_sum = e ** (np.sum(x) - 1)
    
    def calculate_for_axis(axis):
        return -e ** (-float(x[axis])) + exp_sum
    
    if axis is not None:
        return calculate_for_axis(axis)
    else:
        values = np.matrix(np.zeros((len(x), 1)))
        for axis in range(len(x)):
            values[axis] = calculate_for_axis(axis)
    return values

def hessian_f2(x):
    exp_sum = e ** (np.sum(x) - 1)
    hessian = np.matrix(np.zeros((len(x), len(x))))
    for i in range(len(x)):
        for j in range(len(x)):
            hessian[i, j] = e ** (-float(x[i])) * (i == j) + exp_sum
    return hessian

f2P = lambda x: f2(P * x)
grad_f2P = lambda x, axis=None: grad_f2(P * x, axis=axis)
hessian_f2P = lambda x: hessian_f2(P * x)
```

Функции генерации начальной точки и запуска итераций метода с критериями остановки.

```python
def get_start_point2(dim, scale=0.01):
    return np.matrix(sps.uniform.rvs(size=dim, scale=scale)).T

def run_iterations2(method, f, grad_f, start_point=get_start_point2(2), 
                    n_iter=10, time_limit=None, accuracy=None, fmin=None, grad_max=10 ** 7):
    points = [start_point]
    iteration_number = 0
    t = time()
    while True:
        try:
            norm_grad = np.linalg.norm(grad_f(points[-1]))
            if (n_iter is not None and iteration_number >= n_iter) \
               or (time_limit is not None and time() - t >= time_limit) \
               or (accuracy is not None and norm_grad / dim <= accuracy) \
               or (fmin is not None and (f(points[-1] - fmin)) / dim <= accuracy) \
               or math.isnan(norm_grad) \
               or norm_grad / dim > grad_max:
                    break

            points.append(method(points[-1]))
        except OverflowError or LinAlgError:
            break
    
    return points, time() - t
```
