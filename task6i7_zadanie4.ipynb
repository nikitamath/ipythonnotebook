## Гистограммы

### Выбор $h$

Подберем ширину ячейки, используя перекрестную проверку. На график обращать внимание не нужно --- он создается автоматически.

```python
H = []
cv_err = []
n = len(RI)
for bins in range(1, 200):
    h = (max_ri - min_ri) / bins
    H += [h]
    r = hist(RI, bins=bins)[0] / n
    cv_err += [2 / ((n - 1) * h) - ((n + 1) / ((n - 1) * h)) * sum(r ** 2)]
```

Графики величины критерия для разных $h$.

```python
plt.figure(figsize=(17,6))
plt.plot(H, cv_err, 'bo-')
plt.xlim((0, 23))
plt.show()
plt.figure(figsize=(17,6))
plt.plot(H[10:], cv_err[10:], 'bo-')
plt.xlim((0.1, 2.13))
plt.show()
```

Оптимальная ширина ячейки

```python
h_optim = sorted(zip(H, cv_err), key=lambda x: x[1])[0][0]
h_small = 0.1
h_big = 2
print h_optim
```

### Некоторые графики

```python
figure(figsize=(17, 4))
subplot(1, 3, 1), hist(RI, bins=int((max_ri - min_ri) / h_small), normed=True)
subplot(1, 3, 2), hist(RI, bins=int((max_ri - min_ri) / h_optim), normed=True)
subplot(1, 3, 3), hist(RI, bins=int((max_ri - min_ri) / h_big), normed=True)
```

## Ядерные оценки

### Выбор $h$

Подберем ширину ядра, используя перекрестную проверку

```python
def gauss_kernel(x):
    return e ** (- x ** 2 / 2.) / sqrt(2 * pi)
def k_zv(x):
    return gauss_kernel(x / sqrt(2.)) / sqrt(2.) - 2 * gauss_kernel(x)
```

```python
H = np.linspace(0.01, 1.5, 100)
cv_err = []
for h in H:
    s = sum([k_zv((RI[i] - RI[j]) / h) for i in range(n) for j in range(n)])
    cv_err += [s / (h * n ** 2) + 2 * gauss_kernel(0) / (n * h)]
```

График критерия в зависимости от $h$

```python
plt.figure(figsize=(17,6))
plt.plot(H, cv_err, 'ro-')
plt.show()
```

Оптимальная ширина ядра

```python
h_optim = sorted(zip(H, cv_err), key=lambda x: x[1])[0][0]
h_small = 0.05
h_big = 1.2
print h_optim
```

### Некоторые графики

```python
figure(figsize=(18,4))
x = np.atleast_2d(np.linspace(-10, 15, 2000)).T
kde = KernelDensity(kernel='gaussian', bandwidth=h_big).fit(np.atleast_2d(transpose(RI)).T)
subplot(1, 3, 1), plot(x, e ** kde.score_samples(x), 'b-')
kde = KernelDensity(kernel='gaussian', bandwidth=h_optim).fit(np.atleast_2d(transpose(RI)).T)
subplot(1, 3, 2), plot(x, e ** kde.score_samples(x), 'b-')
kde = KernelDensity(kernel='gaussian', bandwidth=h_small).fit(np.atleast_2d(transpose(RI)).T)
subplot(1, 3, 3), plot(x, e ** kde.score_samples(x), 'b-')
```

### Построение доверительного множества

```python
def se(x):
    y = [gauss_kernel((x - RI[i]) / h_optim) / h_optim for i in range(n)]
    s = var(y) * n / (n - 1)
    return sqrt(s / n)
```

```python
m = (15 - (-10)) / (3 * h_optim)
q = norm.ppf((1 + (1 - 0.95) ** (1 / m)) / 2.)
q
```

```python
X = np.atleast_2d(np.linspace(-10, 15, 2000)).T
SE = array([se(x) for x in X])
```

Доверительное множество между зеленой и красной линией. Светло-синяя линия --- сама оценка плотности.

```python
plt.figure(figsize=(17,5))
kde = KernelDensity(kernel='gaussian', bandwidth=h_optim).fit(np.atleast_2d(transpose(RI)).T)
u = e ** kde.score_samples(X) + q * SE
f = e ** kde.score_samples(X)
l = e ** kde.score_samples(X) - q * SE
l = [max(ldfv, 0) for ldfv in l]
plt.plot(X, u, 'g-')
plt.plot(X, f, 'b-', alpha=0.2)
plt.plot(X, l, 'r-')
plt.show()
```
